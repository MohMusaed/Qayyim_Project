```python
# Load the Qwen2.5-1.5B model and tokenizer
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline

# Load the tokenizer and model for Qwen2.5-1.5B
tokenizer = AutoTokenizer.from_pretrained("Qwen/Qwen2.5-1.5B")
model = AutoModelForCausalLM.from_pretrained("Qwen/Qwen2.5-1.5B")

# Define a text generation pipeline using the loaded model and tokenizer
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer)

```

```python
AccDes="""
**الطرف الأول (يسار الصورة):** سيارة سيدان فضية اللون.

**الطرف الثاني (يمين الصورة):** سيارة سيدان سوداء اللون.

**وصف الحادث:**

يبدو أن الحادث عبارة عن تصادم من الخلف، حيث اصطدمت السيارة الفضية (الطرف الأول) بمؤخرة السيارة السوداء (الطرف الثاني).

**الأضرار:**

* **السيارة الفضية:**  يظهر  ضرر كبير في مقدمة السيارة، بما في ذلك غطاء المحرك، والمصد الأمامي، والشبك الأمامي، والمصابيح الأمامية.
"""
```

```python
PartoneDes="""
تم توقف الشارع بشكل مفاجئ وصدمت الذي امامي
"""
```

```python
ParttwoDes="""
تم صدمي من الخلف بعد توقف الشارع بسبب الزحمه
"""
```

```python
# Define the user message or prompt for conversation
prompt = (
    f"""
    يوجد حادث مروري لسيارتين.

    **معلومات الحادث:**
    مدخل إليك وصف الحادث من خلال صورة تم التقاطها: {AccDes}.

    **استمارة تسجيل الطرف الأول:**
    {VecReg1}

    **استمارة تسجيل الطرف الثاني:**
    {VecReg2}

    **وصف الطرف الأول:**
    {PartoneDes}

    **وصف الطرف الثاني:**
    {ParttwoDes}

    **المطلوب:**
    اكتب تقريرا مرتبا وعالي الجودة يحتوي على المعلومات التالية:
    1. **وصف الحادث** بناء على الصورة ووصف الأطراف.
    2. **تعيين الخطأ** ووصف المسؤولية.
    3. **الأضرار المادية لكل سيارة** (بناءً على الصورة ووصف الأطراف).
    4. **تعيين الخطأ بالنسبة المئوية** [100%, 75%, 50%, 25%, 0%].

    إذا كان هناك تعارض في وصف الأطراف، اكتب: "لا يمكن تعيين الخطأ وصفا ويجب التحقق من الأطراف المعنية (نجم أو المرور)".
    """
)

# Generate a response using the pipeline
output = pipe(prompt, max_length=1500, num_return_sequences=1)


```
