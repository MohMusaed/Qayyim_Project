```python
from ultralytics import YOLO

model =YOLO('yolov8n.pt')
```

```python
# إنشاء ملف data.yaml باستخدام Python
# Replace placeholders with actual paths
data_yaml = f"""
train: /content/train
val: /content/valid

nc: 4  # عدد الفئات
names: ['front', 'right_side', 'left_side', 'rear']  # أسماء الفئات
"""

# حفظ ملف data.yaml في النظام
with open('/content/data.yaml', 'w', encoding='utf-8') as file: # Added encoding='utf-8' to support Arabic characters
    file.write(data_yaml)

print("data.yaml file created!")
```

```python
# Set the path to your dataset
dataset_path = '/content/data.yaml'  # Assume the dataset YAML file is available here

# Train the model
model.train(data=dataset_path, epochs=25, imgsz=640)
```

```python
# تحديد مسار الصورة
image_path = '/content/test/images/130_jpg.rf.051fa6e99cab053f8b939dcefc637445.jpg'  # عدل هذا المسار حسب موقع الصورة

# استدلال على الصورة باستخدام YOLOv8
results = model.predict(source=image_path, save=True)

# عرض النتائج (الصورة مع مربعات الإحاطة)
results[0].plot()  # عرض الصورة مع تحديد مكان الصدمة

# طباعة النتائج النصية لتحديد المكان (أمام، جانب أيمن، جانب أيسر، خلف)
# for result in results:
#     for box in result.boxes:
#         cls = int(box.cls[0])  # الحصول على الفئة
#         if cls == 0:
#             print("Damage detected: Front")

#         elif cls == 1:
#             print("Damage detected: Right Side")
#         elif cls == 2:
#             print("Damage detected: Left Side")
#         elif cls == 3:
#             print("Damage detected: Rear")
```
